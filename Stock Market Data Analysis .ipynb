## Representing dataset

This step will be done by creating a function which takes datasets as an input and the graph will be plotted between date and high price of a company share's to check its annual growth

Requirement to make function

* Dataset stored in the variable  
* Data will be plotted in the form of line graph
* The visual representation will be represented with the help of .show() function.   

```
def plot_graph(dataset):
  ax = dataset.plot(kind='line',x='Date',y='High',color='red',figsize=(25,5),title = dataset['Symbol'][0])
  plt.show(ax)
```

```
plot_graph(asian_paints)
```
